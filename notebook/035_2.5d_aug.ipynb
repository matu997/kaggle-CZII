```python
class VolumeDatasetSlices(Dataset):
    def __init__(self, vol_data, mask_data, patch_size=32, stride=16, transform=None):
        """
        Args:
            vol_data (numpy.ndarray):  (N, D, H, W)  (: (N, 184, 640, 640))
            mask_data (numpy.ndarray):  (N, D, H, W)  (: (N, 184, 640, 640))
            patch_size (int):  (: 32)
            stride (int):  (: 16)
            transform (callable, optional): MONAITransforms
        """
        self.vol_data = vol_data
        self.mask_data = mask_data
        self.patch_size = patch_size
        self.stride = stride
        self.transform = transform

        self.indices = []
        # 
        for idx in range(len(vol_data)):
            d = vol_data[idx].shape[0]  # (D)
            # 0, stride, 2*stride, ... patch_size
            for start in range(0, d - patch_size + 1, stride):
                #
                start = start + random.randint(-stride//2, stride//2)
                end = start + patch_size
                # (, , ) 
                #
                if np.sum(mask_data[idx][start:end]) > 0:
                    self.indices.append((idx, start, end))

    def __len__(self):
        # 
        return len(self.indices)

    def __getitem__(self, i):
        idx, start, end = self.indices[i]

        # idx ,  [start:end] 
        volume = self.vol_data[idx][start:end]  # (patch_size, H, W)
        mask   = self.mask_data[idx][start:end] # (patch_size, H, W)

        data_dict = {"image": volume, "label": mask}
        if self.transform:
            data_dict = self.transform(data_dict)

        return data_dict["image"], data_dict["label"]

# ---  ---

# 
# ToTensorD  numpy -> torch.Tensor 
transforms = Compose([
    # 1. Tensor
    ToTensorD(keys=["image", "label"]),
    
    # 2. 
    EnsureChannelFirstd(keys=["image", "label"], channel_dim="no_channel"),
    
    # 3. RAS
    Orientationd(keys=["image", "label"], axcodes="RAS"),
    
    # 4. 90
    RandRotate90d(keys=["image", "label"], prob=0.5, spatial_axes=[1, 2]),
    
    # 5.  ( 0 )
    RandFlipd(keys=["image", "label"], prob=0.5, spatial_axis=0),
    
    # 6.  (32, 640, 640) 
    SpatialPadd(keys=["image", "label"], spatial_size=(32, 640, 640), mode="constant"),
    
    # 7. 
    RandAffined(
        keys=["image", "label"],
        prob=0.5,  # 
        rotate_range=(0.1, 0.1, 0.1),  #  (, : X, Y, Z)
        shear_range=(0.1, 0.1, 0.1),   #  (: X, Y, Z)
        scale_range=(0.3, 0.3, 0.3),   #  (: X, Y, Z)
        translate_range=(100, 109, 100),  #  (: X, Y, Z)
        mode=("bilinear", "nearest"),  #  (: bilinear, : nearest)
        padding_mode="border"  # 
    ),
])
```

```python
from torch.optim.lr_scheduler import LambdaLR

# CV
num_epochs = 40
learning_rate = 1e-3
accumulation_steps = 4  # 

optimizer = torch.optim.AdamW(
    net.parameters(),
    amsgrad=False,
    betas=(0.9, 0.999),
    capturable=False,
    differentiable=False,
    eps=1e-08,
    foreach=None,
    fused=None,
    lr=learning_rate,
    weight_decay=0.01
)

warmup_steps = 30
total_steps = num_epochs * len(dataloader)

def lr_lambda(current_step):
    if current_step < warmup_steps:
        return float(current_step) / float(max(1, warmup_steps))  # 
    return 0.5 * (1. + np.cos(np.pi * (current_step - warmup_steps) / (total_steps - warmup_steps)))

scheduler = LambdaLR(optimizer, lr_lambda=lr_lambda)

scaler = torch.cuda.amp.GradScaler(init_scale=4096)

epoch_losses = []
cv_scores = []  # CV
mask_losses = []

global_step = 0

for epoch in range(num_epochs):
    if epoch % 5 == 0:
        optimizer = torch.optim.AdamW(
            net.parameters(),
            amsgrad=False,
            betas=(0.9, 0.999),
            capturable=False,
            differentiable=False,
            eps=1e-08,
            foreach=None,
            fused=None,
            lr=learning_rate,
            weight_decay=0.01
        )

    print(f" {epoch + 1}/{num_epochs}")
    print("-" * 10)

    net.train()
    running_loss = 0.0

    for i, (volumes, masks) in enumerate(dataloader):
        # 1
        volumes = volumes.squeeze(0)
        masks = masks.squeeze(0)

        batch, D, H, W = volumes.shape

        # 
        volumes = np.pad(volumes, [[0, 0], [0, 0], [0, 640 - H], [0, 640 - W]], mode='constant', constant_values=0)
        masks = np.pad(masks, [[0, 0], [0, 0], [0, 640 - H], [0, 640 - W]], mode='constant', constant_values=0)

        #print(np.sum(masks))
        if np.sum(masks) == 0:
            continue

        batch = {
            'image': torch.from_numpy(volumes).to(device).float(),
            'mask': torch.from_numpy(masks).to(device).long(),
        }

        #
        train_batch = batch

        optimizer.zero_grad(set_to_none=True) if i % accumulation_steps == 0 else None

        with torch.amp.autocast(device_type="cuda", enabled=True):
            output = net(batch)
            loss = output['mask_loss'] / accumulation_steps  # 

        scaler.scale(loss).backward()

        if (i + 1) % accumulation_steps == 0 or (i + 1) == len(dataloader):
            scaler.unscale_(optimizer)
            nn.utils.clip_grad_norm_(net.parameters(), 1.0)
            scaler.step(optimizer)
            scaler.update()
            scheduler.step()
            global_step += 1

        running_loss += loss.item() * accumulation_steps  # 

    epoch_loss = running_loss / len(dataloader)
    epoch_losses.append(epoch_loss)
    print(f"_loss: {epoch_loss:.4f}")

    # CV
    gb, lb_score, mask_loss = evaluate_cv(net, test_id, cfg, valid_dir, device, mask_size)  # test_id: ID
    cv_scores.append(lb_score)
    mask_losses.append(np.mean(mask_loss))
    print(f"CV: lb_score: {lb_score:.4f}")

# 
print("\n:")
for epoch, (loss, lb_score) in enumerate(zip(epoch_losses, cv_scores), start=1):
    print(f" {epoch}: : {loss:.4f}, lb_score: {lb_score:.4f}")
```
